```python
start_time = time.time()
train=df
x_train = train.iloc[:, 1:].values.astype('int32') #(6, ' x ', 10000) baris dan kolom ke 1
labels = train.iloc[:, 0].values.astype('int32')
print(uk(x_train)) #(91, ' x ', 270000)
P=labels.shape[0]  #6
```

```python

X = train.iloc[:, 1:].values   #<class 'numpy.ndarray'> (150, ' x ', 4)
y = train.iloc[:, 0].values     #<class 'numpy.ndarray'>  (150,)

Y=[]
c=y.shape[0]
for i in range(c): 
    D=y[i]
    if D==1:
        Y.append('A')
    elif D==2:
        Y.append('B')
    elif D==3:
        Y.append('C')
    elif D==4:
        Y.append('D')
    elif D==5:
        Y.append('E')

y=Y
print(y)
```

### SVM

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.17, random_state=1234123)  #40proses

pd.Series(y_train).value_counts

# define support vector classifier
svm = SVC(kernel='linear', probability=True, random_state=42)

# fit model
svm.fit(X_train, y_train)

# generate predictions
y_pred = svm.predict(X_test)

# calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Model accuracy is: ', accuracy)

correct = 0
total = len(y_test)
for i in range(total):
    predicted = y_pred[i] #np.argmax(y_pred[i])
    test = y_test[i] #np.argmax(y_test[i])
    print(predicted ,'=',test)
    correct = correct + (1 if predicted == test else 0)
print('Akurasi Data Uji/Tes: {:f}'.format(correct/total))
elapsed_time = time.time() - start_time
time.strftime("%H:%M:%S", time.gmtime(elapsed_time))

```
